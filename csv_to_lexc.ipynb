```python
parts = []
root = ''
morphs = {}

with open('razmetka.csv') as f:
    for line in f:
        morline = line.split(';')
        if morline[5] == 'S':
            if morline[2] == 'корень':
                root = (morline[1], morline[2], morline[4])
            else:
                parts.append((morline[1], morline[2], morline[4]))
            if morline[3] == '1':
                if root:
                    for mor in parts:
                        if mor in morphs:
                            morphs[mor].add((root, morline[0]))
                        else:
                            morphs[mor] = set([(root, morline[0])])
                    parts = []
                    root = ''
            
            
for x in sorted(morphs, key=lambda x: len(morphs[x]), reverse=True)[:20]:
    print(x, len(morphs[x]))
# pprint(morphs)
```

# достаём все корни

```python
roots = {}


with open('razmetka.csv') as f:
    for line in f:
        morline = line.split(';')
        if morline[2] == 'корень' and morline[1] not in roots:
            roots[morline[1]] = [morline[4], morline[0]]
        
pprint(roots)
```

```python
with open('lexc/roots.lexc', 'w') as f:
    for key in roots:
        f.write(f'{key}:{key} AFTER_ROOT ;\t\t!{roots[key][0]}\t{roots[key][1]}\n')
```
